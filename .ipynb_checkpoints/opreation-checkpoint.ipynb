```python
from LinearRegression.model import Model
from sklearn import datasets
```

```python
# 加载数据
diabetes = datasets.load_diabetes()
train_X = diabetes.get("data")
train_y = diabetes.get("target")
```

```python
# 实例化模型, 并训练
model = Model(train_X, train_y, learning_rate=1.95, lamda=0)
model.train(100000)
```

```python
# 展示梯度下降迭代信息
import matplotlib.pyplot as plt
epochs = []
losses = []
plt.figure(figsize=(30, 10))
for epoch, loss in model.history:
    epochs.append(epoch)
    losses.append(loss)
xticks = range(0, epochs[-1], 1000)
plt.xticks(xticks, rotation=45)
plt.plot(epochs, losses, 'r.')
plt.show()
```
