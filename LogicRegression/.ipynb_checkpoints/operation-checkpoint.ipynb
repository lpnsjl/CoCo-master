```python
import tensorflow as tf
from model import Model
```

```python

mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1]*x_train.shape[2], 1)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1]*x_test.shape[2], 1)
```

```python
# 实例化模型
model = Model(784, 10, learning_rate=0.003, lamda=0)
```

```python
# 梯度检查
model.gradient_check(x_train[0], y_train[0])
```

```python
# 模型训练
model.train(x_train, y_train, 256, 300)
```

```python
# 模型的训练损失曲线
import matplotlib.pyplot as plt
epochs = []
losses = []
for epoch, loss in model.history:
    epochs.append(epoch)
    losses.append(loss)
xticks = range(epochs[0], epochs[-1], 10)
plt.xticks(xticks, rotation=45)
plt.plot(epochs, losses, 'r')
plt.plot(epochs, losses, 'b.')
plt.show()
```

```python
# 测试集精度
test_accuracy = model.evaluate(x_test, y_test)
print("test_accuracy: {}".format(test_accuracy))
```
